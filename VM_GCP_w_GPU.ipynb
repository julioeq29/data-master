## Upload data directly from the terminal

### Use for ZIP files

```python
gcloud compute scp 
    --project="wagon-bootcamp-290987" 
    --zone="europe-west4-a" '/Users/julioq/Downloads/bdd100k_images_100k.zip' 
    jupyter@tensorflow-2-6-20220304-105308:~/ 
```

### User for folders

https://cloud.google.com/sdk/gcloud/reference/compute/scp

Word recurse

```python
gcloud compute scp 
    --project="my-gcp-project"
    --zone="us-east1-b" 
    --recurse ~/foo-folder/ gcp-instance-name:~/
```

As the epochs go by:
- the algorithm learns and 
- its prediction error (RMSE) on the training set naturally goes down, and
- so does its prediction error on the validation set. 

After a while, `the validation error stops decreasing and actually starts to go back up`. This indicates that the model has started to overfit the training data. With early stop‚Äê ping you just stop training as soon as the validation error reaches the minimum.
